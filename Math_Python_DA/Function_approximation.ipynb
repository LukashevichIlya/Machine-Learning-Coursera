# Задача 2: аппроксимация функции

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

func_x = np.arange(1, 15, 0.1)
func_y = map(f, func_x)
```

```python
f_1 = f(1.)
f_4 = f(4.)
f_8 = f(8.)
f_10 = f(10.)
f_15 = f(15.)
```

## Приближение многочленом первой степени

```python
first_matrix = np.array([[1, 1], [1, 15]])
first_vector = np.array([f_1, f_15])
first_answer = np.linalg.solve(first_matrix, first_vector)
print first_answer
```

```python
%matplotlib inline

first_x = np.arange(0, 15, 0.1)
first_y = first_answer[0] + first_answer[1] * first_x
plt.plot(first_x, first_y)
plt.show()

plt.plot(func_x, func_y)
plt.show()
```

## Приближение многочленом второй степени

```python
second_matrix = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
second_vector = np.array([f_1, f_8, f_15])
second_answer = np.linalg.solve(second_matrix, second_vector)
print second_answer
```

```python
second_x = np.arange(0, 15, 0.1)
second_y = second_answer[0] + second_answer[1] * second_x + second_answer[2] * second_x ** 2
plt.plot(second_x, second_y)
plt.show()

plt.plot(func_x, func_y)
plt.show()
```

## Приближение многочленом третьей степени

```python
third_matrix = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
third_vector = np.array([f_1, f_4, f_10, f_15])
third_answer = np.linalg.solve(third_matrix, third_vector)

print third_answer
```

```python
third_x = np.arange(0, 15, 0.1)
third_y = third_answer[0] + third_answer[1] * third_x + third_answer[2] * third_x ** 2 + third_answer[3] * third_x ** 3
plt.plot(third_x, third_y)
plt.show()

plt.plot(func_x, func_y)
plt.show()
```
